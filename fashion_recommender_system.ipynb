## 1. Importing Necessary Libraries:

```python
import tensorflow
from keras.applications.resnet50 import ResNet50, preprocess_input
from keras.layers import GlobalMaxPooling2D 
import cv2  # OpenCV library for image processing
import numpy as np
from numpy.linalg import norm  # For vector normalization
import os  # Operating system module for directory operations
from tqdm import tqdm  # Progress bar visualization
import pickle  # Module for serialization of Python objects
```

## 2. Loading Pre-trained ResNet50 Model:


```python
# Load the pre-trained ResNet50 model without the top (fully connected) layers
model=ResNet50(weights='imagenet' , include_top=False , input_shape=(224,224,3))
# Freeze the weights of the pre-trained layers
model.trainable=False  
```

```python
# Add GlobalMaxPooling2D layer to reduce dimensionality and extract features
model= tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
model.summary()
```

## 3. Preprocessing and Extracting Features:

```python
# Function to preprocess images and extract features using the pre-trained model
def extract_feature(img_path , model):
    img=cv2.imread(img_path)                       # Read the image using OpenCV
    img=cv2.resize(img,(224,224))                  # Resize image to match model input size
    img= np.array(img)                             # Convert image to NumPy array
    expand_img = np.expand_dims(img,axis=0)        # Add batch dimension
    pre_img=preprocess_input(expand_img)           # Preprocess input for ResNet50 model
    result=model.predict(pre_img).flatten()        # Extract features using the model
    normalization=result/norm(result)              # Normalize the feature vector
    return normalization                           # Return the normalized feature vector
```

```python
extract_feature('dataset/10037.jpg' , model)
```

## 4. Feature Extraction and Saving:

```python
filename=[]                             # List to store image file paths
feature_list=[]                         # List to store extracted feature vectors

# Iterate over images in the 'dataset' directory and extract features
for file in os.listdir('dataset'):     
    filename.append(os.path.join('dataset',file))  #Append file path to filename list
    
```

```python
filename[0:4]
```

```python
# Iterate over each image file and extract feature
for file in tqdm(filename):
    feature_list.append(extract_feature(file ,model))    # Append feature vector to feature list
    
    
```

```python
import pickle

pickle.dump(feature_list ,open('featurevector.pkl' , 'wb'))
pickle.dump(filename ,open('filename.pkl' , 'wb'))
```
